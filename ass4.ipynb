[3번 문제] range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

```python
def frange(val, basic=0.0, step=0.1):
    l=[]
    
    if val < basic :
        start = val
        stop = basic
    else :
        start = basic
        stop = val
    while start < stop :
        l.append(start)
        start += step
        
    return l

print frange(0.5)
print frange(1.0,2.0)
print frange(2.2,4.0,0.5)
```

[6번 문제] 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

```python
def facto(num) :
    if num==1:
        return 1
    return num*facto(num-1)

print facto(5)
```
